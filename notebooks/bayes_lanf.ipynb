## So what *is* the likelihood of LANF activity?

$ P(A) = 0.5 $ := prior for LANF activity

$ P(\sim A) = 1 - P(A) = 0.5 $ := prior for LANF inactivity

$ P(O \, |\, A) = 0.5 $  := probability of observing LANF quake given activity (this is the result of our study)

$ P(\sim O \,|\,  A) = 1 - P(O \, |\, A) = 0.5 $ := probability of *not* observing LANF earthquake given its activity (also result of this study)

$ P(O \, | \sim A) = 0 $ := probability of false-positive LANF observation

$ P( \sim O | \sim A) = 1 $ := probability of not observing a LANF event given their inactivity


$ P(A | \sim O) = \frac{ P(\sim O|A) P(A)}{P(\sim O|A)P(A) + P(\sim O|\sim A)P(\sim A)} $



```python
%pylab inline
%config InlineBackend.figure_format = 'retina'
```

```python
import numpy as np
```

```python
'''priors for LANF activity'''
p_A = np.linspace(0, 1)  # prior for LANF activity
p_nA = 1 - p_A  # prior for LANF inactivity

'''probability of false positive'''
p_OgnA = 0 # probability of false positive (charitable)
p_nOgnA = 1 - p_OgnA # probability of not false positive ...

'''study results'''
p_OgA = 0.5  # probability of observing EQ given LANF activity
p_nOgA = 1 - p_OgA # probability of not observing EQ given LANF activity
```

```python
p_AgnO = p_nOgA * p_A / ( p_nOgA * p_A + p_nOgnA * p_nA)

#p_AgnO
```

```python
plot(p_A, p_AgnO, label='P(A|~O)')
gca().fill_between(p_A, p_A, y2=p_AgnO, color='lightgrey')
xlabel('prior, P(A)')
ylabel('posterior, P(A|~O)')
legend(loc='upper left')
```
